```python
# Dependencies and Setup
import pandas as pd

# Dummy Variables

totStudentsList = []
passingMathList = []
passingReadingList = []
overallPassRates = []

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_df.head()
```

```python
num_of_schools = school_df["school_name"].nunique()
num_of_students = school_df["student_name"].count()

school_groupby = school_df.groupby(["school_name"]).mean()
totBudget = school_groupby["budget"].sum()

averageMath = school_df["math_score"].mean()
averageReading = school_df["reading_score"].mean()

overallPassRate = (averageMath + averageReading)/2

passingMath = school_df.loc[school_df["math_score"] >= 70, :]
passingReading = school_df.loc[school_df["reading_score"] >= 70, :]

numPassingMath = len(passingMath)
percentPassingMath = (numPassingMath/num_of_students)*100


numPassingReading = len(passingReading)
percentPassingReading = (numPassingReading/num_of_students)*100
percentPassingReading

new_dict = {"Total Schools": [num_of_schools],
           "Total Students": [num_of_students],
           "Total Budget": [totBudget],
           "Average Math Score": [averageMath],
           "Average Reading Score": [averageReading],
           "% Passing Math": [percentPassingMath],
           "% Passing Reading": [percentPassingReading],
           "% Overall Passing Rate": [overallPassRate]}

new_df = pd.DataFrame(new_dict)
new_df
```

```python
# Get math average per school
gb_math = school_df.groupby(["school_name", "Student ID"]).math_score.mean()

gb_math = gb_math.unstack()
mathMean = gb_math.mean(axis = 1)

# Get reading average per school
gb_reading = school_df.groupby(["school_name", "Student ID"]).reading_score.mean()

gb_reading = gb_reading.unstack()
readingMean = gb_reading.mean(axis = 1)

# Get the percentage of students passing math per school
gb_math_t = gb_math.transpose()

gb_math_t = gb_math_t.fillna(-2)
for column in gb_math_t:
    
    studentsperSchool = gb_math_t.loc[(gb_math_t[column] > 0), :]
    passingMathperSchool = gb_math_t.loc[(gb_math_t[column] >= 70), :]
    passingMathList.append((len(passingMathperSchool)/len(studentsperSchool))*100)

#Get the percentage of students passing reading per school
gb_reading_t = gb_reading.transpose()

gb_reading_t = gb_reading_t.fillna(-2)
for column in gb_math_t:
    
    studentsperSchool = gb_reading_t.loc[(gb_reading_t[column] > 0), :]
    passingReadingperSchool = gb_reading_t.loc[(gb_reading_t[column] >= 70), :]
    passingReadingList.append((len(passingReadingperSchool)/len(studentsperSchool))*100)

# Get Overall Pass Rate by averaging passingMathList and passingReadingList    
for mathScore in passingMathList:
    indexed = passingMathList.index(mathScore)
    readingScore = passingReadingList[indexed]
    average = (mathScore + readingScore)/2
    overallPassRates.append(average)
    

# Begin creating the Final DataFrame
gb_df = school_df.groupby(["school_name", "type"]).agg({"student_name":"count",
                                                    "budget": "mean",
                                                    "math_score": "mean",
                                                    "reading_score": "mean"})

gb_df = gb_df.rename(columns = {"student_name": "Total Students", "budget": "Total School Budget", "math_score": "Math Average", "reading_score": "Reading Average"})

# But first, gotta get the budget per student
gb_df["per Student Budget"] = gb_df["Total School Budget"]/gb_df["Total Students"]

# Add the passing math, passing reading,  and overall pass rate columns

gb_df["% Passing Math"] = passingMathList
gb_df["% Passing Reading"] = passingReadingList
gb_df["% Overall Passing Rate"] = overallPassRates

# Re-organize to get the columns in a good order
gb_df = gb_df[["Total Students", "Total School Budget", "per Student Budget", "Math Average", "Reading Average", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
gb_df = gb_df.reset_index()
gb_df_ascending = gb_df.sort_values("% Overall Passing Rate")
gb_df_descending = gb_df.sort_values("% Overall Passing Rate", ascending = False)
```

```python
# Calculate the math score by grade per school
gb_math_by_grade = school_df.groupby(["school_name", "grade"]).math_score.mean()
pd.DataFrame(gb_math_by_grade)
gb_math_by_grade = gb_math_by_grade.unstack()

gb_math_by_grade = gb_math_by_grade.reset_index()

gb_math_by_grade = gb_math_by_grade[["school_name", "9th", "10th", "11th", "12th"]]
gb_math_by_grade = gb_math_by_grade.rename(columns = {"school_name": "School Name"})

gb_math_by_grade
```

```python
# Calculate the reading score by grade per school
gb_reading_by_grade = school_df.groupby(["school_name", "grade"]).reading_score.mean()
pd.DataFrame(gb_reading_by_grade)
gb_reading_by_grade = gb_reading_by_grade.unstack()

gb_reading_by_grade = gb_reading_by_grade.reset_index()

gb_reading_by_grade = gb_reading_by_grade[["school_name", "9th", "10th", "11th", "12th"]]
gb_reading_by_grade = gb_reading_by_grade.rename(columns = {"school_name": "School Name"})

gb_reading_by_grade.index.names = ['Index']

gb_reading_by_grade
```

```python
# Get the scores by school spending
gb_df_spending = gb_df

spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
gb_df_spending["Spending/Student"] = pd.cut(gb_df_spending["per Student Budget"], spending_bins, labels = group_names)

# Now Group By bins
spending_groupby = gb_df_spending.groupby(["Spending/Student"]).mean()
spending_groupby = spending_groupby[["Math Average", "Reading Average", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
spending_groupby
```

```python
# Get the scores by school spending
gb_df_size = gb_df

size_bins = [0, 1000, 2000, 5000]
group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

gb_df_size["Size"] = pd.cut(gb_df_size["Total Students"], size_bins, labels = group_names)

# Now Group By bins
size_groupby = gb_df_size.groupby(["Size"]).mean()
size_groupby = size_groupby[["Math Average", "Reading Average", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]]
size_groupby
```
